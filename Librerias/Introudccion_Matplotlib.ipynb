# Introducción Matplotlib


```python
# Version de matplotlib
import matplotlib
print("Versión de matplotlib:", matplotlib.__version__)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Configuración para mostrar los gráficos en el notebook
%matplotlib inline
```

## Crear un Gráfico de Línea
Vamos a crear un gráfico de línea simple usando datos generados con NumPy.

```python
# Generar datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear el gráfico de línea
plt.plot(x, y)
plt.show()
```

## Personalizar el Gráfico
Ahora personalizaremos el gráfico cambiando colores, estilos y tamaños.

```python
# Configurar el tamaño de la figura
plt.figure(figsize=(10, 6))

# Crear múltiples líneas con diferentes estilos
plt.plot(x, np.sin(x), 'r--', label='sin(x)')
plt.plot(x, np.cos(x), 'b-', label='cos(x)')

# Mostrar el gráfico
plt.show()
```

## Agregar Etiquetas y Leyendas
Añadiremos etiquetas a los ejes y una leyenda para hacer el gráfico más informativo.

```python
plt.figure(figsize=(10, 6))
plt.plot(x, np.sin(x), 'r--', label='sin(x)')
plt.plot(x, np.cos(x), 'b-', label='cos(x)')

# Agregar títulos y etiquetas
plt.title('Funciones Trigonométricas')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)

plt.show()
```

## Guardar el Gráfico como Imagen
Finalmente, aprenderemos cómo guardar nuestro gráfico como un archivo de imagen.

```python
# Crear el gráfico final
plt.figure(figsize=(10, 6))
plt.plot(x, np.sin(x), 'r--', label='sin(x)')
plt.plot(x, np.cos(x), 'b-', label='cos(x)')
plt.title('Funciones Trigonométricas')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)

# Guardar el gráfico
plt.savefig('funciones_trigonometricas.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Trabajando con Subplots
Los subplots nos permiten crear múltiples gráficos en una sola figura. Matplotlib ofrece varias formas de crear subplots:

```python
# Crear una figura con 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Primer subplot: Función seno
axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title('Función Seno')

# Segundo subplot: Función coseno
axs[0, 1].plot(x, np.cos(x), 'g-')
axs[0, 1].set_title('Función Coseno')

# Tercer subplot: Función tangente
axs[1, 0].plot(x, np.tan(x), 'r-')
axs[1, 0].set_ylim(-5, 5)  # Limitamos el eje y para mejor visualización
axs[1, 0].set_title('Función Tangente')

# Cuarto subplot: Todas las funciones juntas
axs[1, 1].plot(x, np.sin(x), 'b-', label='sin(x)')
axs[1, 1].plot(x, np.cos(x), 'g-', label='cos(x)')
axs[1, 1].plot(x, np.tan(x), 'r-', label='tan(x)')
axs[1, 1].set_ylim(-5, 5)
axs[1, 1].legend()
axs[1, 1].set_title('Todas las Funciones')

# Ajustar el espaciado entre subplots
plt.tight_layout()
plt.show()
```

### Subplots con Diferentes Tamaños
También podemos crear layouts más complejos usando `GridSpec`:

```python
from matplotlib.gridspec import GridSpec

# Crear una figura con un layout personalizado
fig = plt.figure(figsize=(12, 8))
gs = GridSpec(2, 3, figure=fig)

# Crear subplot que ocupa toda la primera fila
ax1 = fig.add_subplot(gs[0, :])
ax1.plot(x, np.sin(2*x))
ax1.set_title('Gráfico Principal')

# Crear tres subplots en la segunda fila
ax2 = fig.add_subplot(gs[1, 0])
ax2.plot(x, np.sin(x), 'r')
ax2.set_title('Subplot 1')

ax3 = fig.add_subplot(gs[1, 1])
ax3.plot(x, np.cos(x), 'g')
ax3.set_title('Subplot 2')

ax4 = fig.add_subplot(gs[1, 2])
ax4.plot(x, np.tan(x), 'b')
ax4.set_ylim(-5, 5)
ax4.set_title('Subplot 3')

plt.tight_layout()
plt.show()
```
